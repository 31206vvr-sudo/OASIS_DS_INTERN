```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\31206\Downloads\Unemployment in India.csv")
df.columns = [col.strip() for col in df.columns]
df.dropna(inplace=True)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dropna(subset=['Date'], inplace=True)
```

```python
print("✅ Dataset shape:", df.shape)
print("✅ Columns:", df.columns.tolist())
print("✅ Unique regions:", df['Region'].nunique())
```

```python
states_to_plot = ['Andhra Pradesh', 'Bihar', 'Delhi', 'Karnataka']
plt.figure(figsize=(12, 6))
for state in states_to_plot:
    state_data = df[df['Region'] == state]
    sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=state_data, label=state)

plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.tight_layout()
plt.show()
```
